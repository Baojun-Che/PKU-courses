```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
```

```python
def plot_scatter_from_file(filename, output_filename=None, cmap='viridis'):
    """
    绘制散点图对比图：左边是原始数据，右边是参考函数（如果提供）
    """
    # 读取数据
    data = np.loadtxt(filename)
    x = data[:, 0]
    y = data[:, 1]
    u_num = data[:, 2]
    u_ref = data[:, 3]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    axes = [ax1, ax2]
    
    # 左边：原始数据
    scatter1 = axes[0].scatter(x, y, c=u_num, cmap=cmap, s=10, alpha=1, 
                             edgecolors='black', linewidth=0.5)
    plt.colorbar(scatter1, ax=axes[0])
    axes[0].set_xlabel('x')
    axes[0].set_ylabel('y')
    axes[0].set_title('Numercial Solution')
    axes[0].grid(True, alpha=0.3)
    
    scatter2 = axes[1].scatter(x, y, c=u_ref, cmap=cmap, s=10, alpha=1, 
                             edgecolors='black', linewidth=0.5)
    
    plt.colorbar(scatter2, ax=axes[1])
    axes[1].set_xlabel('x')
    axes[1].set_ylabel('y')
    axes[1].set_title('Reference Solution')
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    
    if output_filename:
        plt.savefig(output_filename, dpi=300, bbox_inches='tight')
        print(f"对比图已保存为: {output_filename}")
    
    plt.show()
```

```python
input_file = "../results/explicit_solver.txt"

PI = 3.14159265

try:      
    # 同时绘制散点图作为对比
    print("\n正在绘制散点图...")
    plot_scatter_from_file(input_file, "explicit_solver.png")
        
except FileNotFoundError:
    print(f"错误: 找不到文件 {input_file}")
    print("请确保数据文件存在于当前目录")
except Exception as e:
    print(f"处理过程中出错: {e}")
```
